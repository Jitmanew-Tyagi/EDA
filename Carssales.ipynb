```python
columns = [ 'Sales_in_thousands', '__year_resale_value', 'Price_in_thousands', 'Engine_size', 'Horsepower','Wheelbase', 'Width', 'Length', 'Curb_weight', 'Fuel_capacity',
       'Fuel_efficiency',  'Power_perf_factor']
```

```python
def impute_median(ser):
    return ser.fillna(ser.median())
for i in columns:
    data[i] = data[i].transform(impute_median)
```

```python
data.isnull().sum()
```

```python
sns.regplot(x=data['Price_in_thousands'],y=data['Horsepower'])
```

```python
df_num = data.select_dtypes(include = ['float64', 'int64'])
for i in range(0, len(df_num.columns),5):
    sns.pairplot(df_num, y_vars ='Price_in_thousands', x_vars = df_num.columns[i:i+5]  )
```
